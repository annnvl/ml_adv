```python
#0,1
train = dep[:int(dep.size*0.8)]
test = dep[int(dep.size*0.8):]
train.plot(figsize=(15, 5))
```

```python
#2
from copy import deepcopy
train_boxcox = deepcopy(train)
#delete zeros
train_boxcox[train['item 29673'] ==0] = np.ones_like(train_boxcox[train['item 29673'] ==0]) * .0001
train_boxcox['item 29673'] = stats.boxcox(train_boxcox['item 29673'])[0]
train_boxcox['item 29673'].plot(figsize=(15, 5))
```

```python
kpss_stat= stattools.kpss(train_boxcox['item 29673'])
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

It is stationary

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(train_boxcox.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(train_boxcox.values.squeeze(), lags=40, ax=ax2)
```

```python
#q = 1, 2, 3, 5, 8, 9; p = 1, 2, 5, 8
arima_mod98 = sm.tsa.ARIMA(train_boxcox, (9,0,8)).fit()
print(arima_mod98.params)
print(arima_mod98.aic, arima_mod98.bic, arima_mod98.hqic)
```

```python
Q = [1, 2, 3, 5, 8, 9]
P = [1, 2, 5, 8]
arima_mod = {}
aic = {}
for q in Q:
    for p in P:
        if(not(q, p) in [(1, 5), (1, 8), (2, 5), (2, 8), (3, 5), (3, 8), (5, 8)] ):
            arima_mod[str(q)+str(p)] = sm.tsa.ARIMA(train_boxcox, (q,0,p)).fit()
            aic[str(q)+str(p)] = arima_mod[str(q)+str(p)].aic

print(max(aic))
```

```python
#best in 9, 8
plt = arima_mod98.plot_predict()
```

```python
forecast = arima_mod98.predict()
forecast.plot(figsize=(15,5))
```

```python
pd.DataFrame({'lags':range(30), 'pvalue':sm.stats.diagnostic.acorr_ljungbox(arima_mod98.resid.values, lags=30)[1], 
              'critial':np.array([0.05]*30)}).set_index('lags').plot(figsize=(15,5))
```

Values are not correlated

```python
kpss_stat= stattools.kpss(arima_mod98.resid.values)
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

It is stationary

```python
sm.stats.durbin_watson(arima_mod98.resid.values)
```

```python
stat = stats.wilcoxon(arima_mod98.resid.values)
print('Statistic: %f\npvalue: %.2f' %(stat[0], stat[1]))
```

Residuals are not biased!

Model is OK
